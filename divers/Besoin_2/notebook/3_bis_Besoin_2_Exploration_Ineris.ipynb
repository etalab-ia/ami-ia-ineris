# II. Chargement des données

```python
data = pd.read_csv('DB_INERIS.csv', sep=';')

data['Type de point de mesure'].replace({"source":"Source"}, inplace=True)
```

# III. Analyse volumétrique

## III.a Analyse des valeurs nulles

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    print(data.isnull().sum())
```

## III.b Analyse de la catégorie "Fournisseur"

```python
col = "Fournisseur"
f,ax=plt.subplots(1,2,figsize=(18,8))
data[col].value_counts().plot.pie(autopct='%1.1f%%',ax=ax[0],shadow=True)
ax[0].set_title(col)
ax[0].set_ylabel('')
sns.countplot(col,data=data,ax=ax[1])
ax[1].set_title(col)
ax[1].tick_params(labelrotation=45)
plt.show()
```
